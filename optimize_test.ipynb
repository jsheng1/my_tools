```python
#from torch import quantile
import optuna
import numpy as np
from scipy.stats import beta
k2=np.array([0.99, 0.995])
quantile_target = 0.001 + np.sqrt(1/(1-k2))*0.005
def objective(trial):
    a = trial.suggest_float("a", 1e-4, 1e-2, log=True)
    b = trial.suggest_float("b", 1e1, 1e3, log=True)
    quantile = beta.ppf(k2, a, b)
    error = ((quantile - quantile_target) / quantile_target) **2
    return np.sum(error)

study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=100, )

a_opt = study.best_params["a"]
b_opt = study.best_params["b"]

print(a_opt, b_opt)
print(beta.ppf(k2, a_opt, b_opt))
print(quantile_target)
```
